## Objective
Validate calculated KPIs against business rules and identify inconsistencies.

```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/production_records.csv')
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Validate OEE calculation
print('OEE Statistics:')
print(f'Mean OEE: {df["OEE_Percent"].mean():.2f}%')
print(f'Median OEE: {df["OEE_Percent"].median():.2f}%')
print(f'Target OEE: 85%')
print(f'Records meeting target: {(df["OEE_Percent"] >= 85).sum()} / {len(df)}')
```

```python
# Defect rate validation
print('\nDefect Rate Statistics:')
print(f'Mean Defect Rate: {df["Defect_Rate_Percent"].mean():.2f}%')
print(f'Target: ≤ 2%')
print(f'Records exceeding target: {(df["Defect_Rate_Percent"] > 2).sum()} / {len(df)}')
```

```python
# Production line performance comparison
line_performance = df.groupby('Production_Line').agg({
    'OEE_Percent': 'mean',
    'Defect_Rate_Percent': 'mean',
    'Downtime_Hours': 'sum',
    'Production_Volume': 'sum'
}).round(2)

print('\nProduction Line Performance:')
print(line_performance)
```
