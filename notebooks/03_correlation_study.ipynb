## Objective
Analyze the relationship between downtime and defect rates to identify operational insights.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

df = pd.read_csv('../data/production_records.csv')
```

```python
# Calculate correlation
correlation = df['Downtime_Hours'].corr(df['Defect_Rate_Percent'])
print(f'Correlation coefficient: {correlation:.3f}')

# Perform statistical test
r, p_value = stats.pearsonr(df['Downtime_Hours'], df['Defect_Rate_Percent'])
print(f'Pearson r: {r:.3f}')
print(f'P-value: {p_value:.4f}')
print(f'Significant: {"Yes" if p_value < 0.05 else "No"}')
```

```python
# Scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='Downtime_Hours', y='Defect_Rate_Percent', data=df, 
            scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Downtime vs Defect Rate Correlation')
plt.xlabel('Downtime (hours)')
plt.ylabel('Defect Rate (%)')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Correlation matrix
corr_cols = ['OEE_Percent', 'Defect_Rate_Percent', 'Downtime_Hours', 
             'Production_Volume', 'Throughput_Efficiency_Percent']
corr_matrix = df[corr_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, linewidths=1)
plt.title('KPI Correlation Matrix')
plt.show()
```
